```python
import sys
sys.path.append('/user/mahaohui/IMG2SMILES/decimer_transformer')

from DECIMER import predict_SMILES
```

```python
import numpy as np
import pandas as pd
import os
import uuid
import glob


root_dir = "/user/mahaohui/data/IMG2SMILES/example_pictures"
dir_list = os.listdir(root_dir)
print(dir_list)

picture_data = {'url': [], 'SMILES': [], 'picture_id': [], "paper_id": []}
picture_df = pd.DataFrame(picture_data)
molecular_data = {"SMILES": []}
molecular_df = pd.DataFrame(molecular_data)
molecular_picture_data = {"SMILES": [], "picture_id": []}
molecular_picture_df = pd.DataFrame(molecular_picture_data)

for subdir in dir_list:
    for png_path in glob.glob(os.path.join(root_dir,subdir,"segments","*bnw*.png")):
        SMILES = predict_SMILES(png_path)
        print(png_path, SMILES)
        
        picture_id = str(uuid.uuid4())
        m_picture_row = {'url': png_path, 'SMILES': SMILES, 'picture_id': picture_id, "paper_id":png_path.split("/")[-3]}
        picture_df.loc[len(picture_df.index)] = m_picture_row
        
        molecular_df_row = {'SMILES': SMILES}
        molecular_df.loc[len(molecular_df.index)] = molecular_df_row
        
        molecular_picture_row = {'SMILES': SMILES, 'picture_id': picture_id}
        molecular_picture_df.loc[len(molecular_picture_df.index)] = molecular_picture_row
```

```python
print(picture_df)
picture_df.to_csv("./example_pdfs/m_picture.csv",index=False)
```

```python
print(molecular_df)
molecular_df.to_csv("./example_pdfs/molecular.csv",index=False)
```
