```python
import os
from PIL import Image
import ipyplot

from decimer_image_classifier import DecimerImageClassifier
```

```python
decimer_classifier = DecimerImageClassifier()
```

```python
import shutil

test_dir = '/user/mahaohui/IMG2SMILES/collect_NCBI/example_pictures/da569c97cc3cb686a19b/segments'

#images.remove('MMU298754.jpg')
#images.remove('CNP0460730.1_2.png')
def classify_imgs(test_dir):
    images = os.listdir(test_dir)
    images_loaded=[]
    if not os.path.exist(test_dir+"/not_molecular"):
        os.mkdir(test_dir+"/not_molecular")
    for img_path in images:
        img = Image.open(os.path.join(test_dir, img_path))
        # score = decimer_classifier.get_classifier_score(img)
        result = decimer_classifier.is_chemical_structure(img)
        images_loaded.append(img)
        print(result)
        if result == False:
            shutil.move(img_path,test_dir+"not_molecular")
        
classify_imgs(test_dir)
```

```python
labels = []
for img in images_loaded:
    result = decimer_classifier.is_chemical_structure(img)
    print(result)
    labels.append(result)
```

```python
ipyplot.plot_images(images_loaded, labels=labels, max_images=20, img_width=100)
```
