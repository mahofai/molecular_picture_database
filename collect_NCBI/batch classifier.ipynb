```python
import shutil

test_dir = '/user/mahaohui/data/IMG2SMILES/example_pictures/da569c97cc3cb686a19b/segments'

#images.remove('MMU298754.jpg')
#images.remove('CNP0460730.1_2.png')
def classify_imgs(test_dir):
    print(test_dir)
    images = os.listdir(test_dir)
    images_loaded=[]
    if not os.path.exists(test_dir+"_not_molecular"):
        os.mkdir(test_dir+"_not_molecular")
    for img_path in images:
        img = Image.open(os.path.join(test_dir, img_path))
        # score = decimer_classifier.get_classifier_score(img)
        result = decimer_classifier.is_chemical_structure(img)
        images_loaded.append(img)
        print(result)
        if result == False:
            shutil.move(os.path.join(test_dir, img_path),test_dir+"_not_molecular/"+img_path)
        
# classify_imgs(test_dir)
```

```python
root_dir = "/user/mahaohui/data/IMG2SMILES/example_pictures"
dir_list = os.listdir(root_dir)
print(dir_list)
for subdir in dir_list:
    classify_imgs(str(os.path.join(root_dir,subdir,"segments")))
    print(f"finish classification on {subdir}")
```
